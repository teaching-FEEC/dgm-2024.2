```python
import pandas as pd
import matplotlib.pyplot as plt
import wandb

KEYS = ['g1_train_loss', 'g1_val_loss', 'd1_train_loss', 'd1_val_loss', 'g2_train_loss', 'g2_val_loss', 'd2_train_loss', 'd2_val_loss']
```

```python
# loading losses from wandb
api = wandb.Api()
run = api.run(f"oficina/satellite2map_cyclegan/ytlw2mg1")
history = run.scan_history(keys=KEYS)

losses = {}
for key in KEYS:
    losses[key] = [row[key] for row in history]

```

# Comparing discriminators losses

```python
disc1_train_loss = losses['d1_train_loss']
disc1_val_loss = losses['d1_val_loss']
disc2_train_loss = losses['d2_train_loss']
disc2_val_loss = losses['d2_val_loss']

plt.figure()
plt.plot(disc1_train_loss, '-', label='disc1_train')
plt.plot(disc1_val_loss,   '--', label='disc1_val')
plt.plot(disc2_train_loss, '-', label='disc2_train')
plt.plot(disc2_val_loss,   '--', label='disc2_val')
plt.legend()
plt.grid()
plt.title('Discriminators Losses')
```

# Comparing generators losses

```python
gen1_train_loss = losses['g1_train_loss']
gen1_val_loss =   losses['g1_val_loss']
gen2_train_loss = losses['g2_train_loss']
gen2_val_loss =   losses['g2_val_loss']

plt.figure()
plt.plot(gen1_train_loss, '-', label='gen1_train')
plt.plot(gen1_val_loss,   '--', label='gen1_val')
plt.plot(gen2_train_loss, '-', label='gen2_train')
plt.plot(gen2_val_loss,   '--', label='gen2_val')
plt.legend()
plt.grid()
plt.title('Generators Losses')
```
