```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
""" Gráficos de Perdas do Gerador e Discriminador do Modelo Pix2Pix """

gen_train_loss = r'C:\Users\cosme\Meu Drive\2-Estudos\5-Doutorado\IA376 - IA Generativa\Projeto\generator_train_loss.csv'
gen_val_loss = r'C:\Users\cosme\Meu Drive\2-Estudos\5-Doutorado\IA376 - IA Generativa\Projeto\generator_val_loss.csv'
dis_train_loss = r'C:\Users\cosme\Meu Drive\2-Estudos\5-Doutorado\IA376 - IA Generativa\Projeto\discriminator_train_loss.csv'
dis_val_loss = r'C:\Users\cosme\Meu Drive\2-Estudos\5-Doutorado\IA376 - IA Generativa\Projeto\discriminator_val_loss.csv'
```

```python
df_gen_train_loss = pd.read_csv(gen_train_loss)
df_gen_val_loss = pd.read_csv(gen_val_loss)
df_dis_train_loss = pd.read_csv(dis_train_loss)
df_dis_val_loss = pd.read_csv(dis_val_loss)
```

```python
df_dis_val_loss
```

```python
plt.figure(figsize=(10, 5))
plt.plot(df_gen_train_loss['Step'], df_gen_train_loss['eager-fire-1 - generator_train_loss'], label='Gen Training loss')
plt.plot(df_dis_train_loss['Step'], df_dis_train_loss['eager-fire-1 - discriminator_train_loss'], label='Dis Training loss')
plt.xlabel('Step')
plt.ylabel('Loss')
plt.title('Training Loss x Steps')
plt.legend()
#plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 5))
plt.plot(df_gen_val_loss['Step'], df_gen_val_loss['eager-fire-1 - generator_val_loss'], label='Gen Val loss')
plt.plot(df_dis_val_loss['Step'], df_dis_val_loss['eager-fire-1 - discriminator_val_loss'], label='Dis Val loss')
plt.xlabel('Step')
plt.ylabel('Loss')
plt.title('Validation Loss x Steps')
plt.legend()
#plt.grid(True)
plt.show()
```
